```python
def tokenizer(raw_texts, pos=["Noun","Alpha","Verb","Number"], stopword=[]):
    p = okt.pos(raw_texts, 
            norm=True,   # 정규화(normalization)
            stem=True    # 어간추출(stemming)
            )
    o = [word for word, tag in p if len(word) > 1 and tag in pos and word not in stopword]
    return(o)
```

```python
genres = ['드라마', '판타지', '서부', '공포', '멜로', '로맨스', '모험', '스릴러', '느와르', '컬트', '다큐멘터리', '코미디', '가족', '미스터리', '전쟁', '애니메이션',
          '범죄', '뮤지컬', 'SF', '액션', '무협', '에로', '서스펜스', '서사', '블랙코미디', '실험', '영화카툰', '영화음악', '영화패러디포스터', '공연실황']

genre_scores = dict()
```

```python
for genre in genres:
  mbti_sim_list = []
  for idx, repeat in enumerate(['E', 'I', 'S', 'N', 'F', 'T', 'P', 'J']):
    temp_sim_list = []
    for word in mbti_words[idx]:
      try:
        temp_sim_list.append(model.wv.similarity(genre, word))
      except:
        temp_sim_list.append(1e-3)
    mbti_sim_list.append(sum(temp_sim_list) / len(temp_sim_list))
  genre_scores[genre] = mbti_sim_list
```

```python
def cal_dataset_genre_score(genres, genre_scores):
  genre_sum_list = []
  for genre in genres:
    genre_sum_list += genre_scores[genre]
  genre_total_score = np.array(genre_sum_list) / len(genres)
  return genre_total_score
```

```python
movie['genre_score'] = movie['genre_split'].apply(lambda genres: cal_dataset_genre_score(genres, genre_scores = genre_scores))
```
